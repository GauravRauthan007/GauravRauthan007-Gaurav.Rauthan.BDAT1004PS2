Function fileLength(), given to you, takes the name of a file as input and returns the length of the file. if the file cannot be found by the interpreter or if it cannot be read as a text file, an exception will be raised. Modify function fileLength() so that a friendly message is printed instead.



```python
def fileLength(file_name):
    try:
        with open(file_name) as file: # use with statement to open and close the file
            contents = file.read() # read the file content as a string
            print(len(contents)) # print length of the string
    except FileNotFoundError: # if the file is not found
        print(f'File {file_name} not found.') # print a message
```

Write a class named Marsupial that can be used as shown below:

m = Marsupial()

m.put_in_pouch('doll')

m.put_in_pouch('firetruck')

m.put_in_pouch('kitten')

m.pouch_contents()

['doll', 'firetruck', 'kitten']



```python
class Marsupial:
    def __init__(self):
        self.pouch = [] # make an empty list to store the items

    def put_in_pouch(self, item):
        self.pouch.append(item) # add the item to the end of the list

    def pouch_contents(self):
        return self.pouch # return the list of items
```

```python
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()
```

Create a new class called "Kangaroo" that's like a special type of "Marsupial." The Kangaroo class should do these things:

a. When you create a Kangaroo, you need to tell it where it is on a grid. So, you'll need to give it two numbers, like the x and y coordinates.

b. Kangaroos should be able to jump. You can tell a Kangaroo to jump by giving it two numbers: how far to move left or right (along the x-axis) and how far to move up or down (along the y-axis).

c. When you want to print out a Kangaroo, it should say where it is. For example, if you have a Kangaroo at coordinates (0, 0), and you print it, it should say, "I am a Kangaroo located at coordinates (0, 0)."

Here's an example of how you can use the Kangaroo class:

```python
class Kangaroo(Marsupial): # define Kangaroo as a subclass of Marsupial
    def __init__(self, x, y):
        Marsupial.__init__(self) # call the Marsupial constructor to initialize the pouch attribute
        self.x = x # establish x coordinate to an attribute
        self.y = y # establish y coordinate to an attribute

    def jump(self, dx, dy):
        self.x += dx # update x coordinate by adding dx
        self.y += dy # update y coordinate by adding dy

    def __str__(self):
        return f'I am a Kangaroo located at coordinates ({self.x},{self.y})' # return a string representation of the object
```

```python
def collatz(x):
    print(x) # print value of x
    if x == 1: # base case: if x is 1, stop the Repetition
        return
    elif x % 2 == 0: # Repetition case: if x is even, divide it by 2 and call collatz again
        collatz(x // 2)
    else: # Repetition case: if x is odd, multiply it by 3 and add 1 and call collatz again
        collatz(3 * x + 1)
```

Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.

```python
def binary(n):
    if n > 1:
        binary(n // 2) # Repetition call with integer division by 2
    print(n % 2, end="") # print the remainder (0 or 1) without a newline
```

# Question 6

Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.

infile = open('w3c.html')

content = infile.read()

infile.close()

hp = HeadingParser()

hp.feed(content)

W3C Mission Principles



```python
from html.parser import HTMLParser # import the HTMLParser class from the standard library module html.parser
import re # import the re module for regular expressions

class HeadingParser(HTMLParser): # define HeadingParser as a subclass of HTMLParser
    def __init__(self):
        HTMLParser.__init__(self) # call the HTMLParser constructor to initialize the base class attributes
        self.indent = 0 # create an attribute to store the current indentation level

    def handle_starttag(self, tag, attrs):
        if re.match('h[1-6]', tag): # check if the tag is a valid heading tag (h1, h2, etc.) using a regular expression
            self.indent = int(tag[1]) - 1 # update the indentation level based on the tag number

    def handle_data(self, data):
        if self.indent > 0: # check if the indentation level is positive
            print(' ' * self.indent + data) # print the data with the appropriate indentation
```

```python
infile = open('w3c.txt') # open the file w3c.html
content = infile.read() # read the file content as a string
infile.close() # close the file
hp = HeadingParser() # create a HeadingParser object
hp.feed(content) # feed the content to the parser
```

Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.

webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)

http://reed.cs.depaul.edu/lperkovic/csc242/test1.html

http://reed.cs.depaul.edu/lperkovic/csc242/test2.html

http://reed.cs.depaul.edu/lperkovic/csc242/test4.html

http://reed.cs.depaul.edu/lperkovic/csc242/test3.html

http://reed.cs.depaul.edu/lperkovic/csc242/test4.html



```python
import urllib.request # import the module for requesting URLs
import re # import the module for regular expressions

def webdir(url, depth, indent):
    print(' ' * indent + url) # print the current URL with the appropriate indentation
    if depth == 0: # base case: if depth is 0, stop the recursion
        return
    else: # recursive case: if depth is positive, visit the links on the current page
        try:
            response = urllib.request.urlopen(url) # try to open the URL
            html = response.read().decode() # read the HTML content as a string
            response.close() # close the response object
            links = re.findall('href="(.*?)"', html) # find all the links on the page using a regular expression
            for link in links: # loop through each link
                if link.startswith('http'): # check if the link is a valid URL
                    webdir(link, depth - 1, indent + 1) # call webdir again with the link, reduced depth, and increased indent
        except: # if an error occurs while opening or reading the URL
            print(' ' * indent + 'Error: could not open or read ' + url) # print an error message with the appropriate indentation
```

# Question 9

Suppose list words is defined as follows:

words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over','the', 'lazy', 'dog']

Write list comprehension expressions that use list words and generate the following lists:

a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE', 'LAZY', 'DOG']

b) ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

c) [3, 5, 5, 3, 5, 4, 3, 4, 3] (the list of lengths of words in list words).

d) [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN', 'brown', 5], ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5], ['OVER', 'over', 4], ['THE', 'the', 3], ['LAZY', 'lazy', 4], ['DOG', 'dog', 3]] (the list containing a list for every word of list words, where each list contains the word in uppercase and lowercase and the length of the word.)

e) ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog'] (the list of words in list words containing 4 or more characters.)



```python
# Original list of words
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# a) Convert all words to uppercase
upper_words = [word.upper() for word in words]
print("Uppercase Words:", upper_words)
```

```python
# b) Convert all words to lowercase
lower_words = [word.lower() for word in words]
print("Lowercase Words:", lower_words)
```

```python
# c) List of lengths of words
word_lengths = [len(word) for word in words]
print("Word Lengths:", word_lengths)
```

```python
# d) List containing word, lowercase word, and length for each word
word_info = [[word.upper(), word.lower(), len(word)] for word in words]
print("Word Info:", word_info)
```

```python
# e) List of words containing 4 or more characters
long_words = [word for word in words if len(word) >= 4]
print("Words with 4 or more characters:", long_words)
```
